```python
X = df[['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

best_silhouette_score = -1
best_k = -1

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_pca, cluster_labels)
    print(f"K: {k} - Silhouette Score: {silhouette_avg}")
    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_k = k

print(f"Best Silhouette Score: {best_silhouette_score}")
print(f"Best K: {best_k}")
```

```python
kmeans_best = KMeans(n_clusters=best_k, random_state=42)
kmeans_best.fit(X_pca)
cluster_labels = kmeans.labels_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=cluster_labels, alpha=0.5)
plt.title(f"Clustering results for best KMeans {best_k}")
plt.xlabel("Principal component 1")
plt.ylabel("Principal component 2")
plt.tight_layout()
plt.colorbar(label="cluster")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
accuracy = mean_squared_error(y_test, y_pred)
print(f"MSE: {accuracy}")
```
